```python
# Guess a number game

from tkinter import *
import random

w = Tk() 
w.title('Guess A Number Game') 
w.geometry('600x400')
# background color for window
w.config(bg='#FFA4A4') 
w.resizable(width=False, height=False)

# main logic

ranNum = random.randint(1, 10) 
chance = 4 
displayResult = StringVar()

def check_guess():
    global ranNum 
    global chance 
    userInput = int(user_input.get()) 
    if chance > 0: 
        if userInput == ranNum: 
            msg = "Congratulation. You Won! " + str(ranNum) + " is the right number"
            user_input.config(state='disabled')
            guess_button.config(state='disabled')
        elif userInput > ranNum: 
            chance = chance - 1 
            msg = "Your Guess Is Too High. Think Of A Smaller Number"
            user_input.delete(0, END)
        elif userInput < ranNum: 
            chance = chance - 1 
            msg = "Your Guess Is Too Low. Think Of A Higher Number"
            user_input.delete(0, END)
    else: 
        msg = "Game Over! " + str(chance) + " Attempt Left. The Correct Number Was " + str(ranNum)
        user_input.config(state='disabled')
        guess_button.config(state='disabled')
        
    displayResult.set(msg)

# Create Widgets 
title = Label(w, text='Guess A Number Game', font=('Times New Roman', 20), fg='#BADFDB', bg='#FFA4A4')
gameInstruction = Label(w, text='Guess The Number Between 1 to 10 in 4 Attempts', font=('Times New Roman', 13), fg='#BADFDB', bg='#FFA4A4')
user_input = Entry(w, font=('Times New Roman', 12))
guess_button = Button(w, text='Guess', font=('Times New Roman', 13), fg='#B6771D', bg='yellow', command=check_guess)
exit_button = Button(w, text='Exit Game', font=('Times New Roman', 14), fg='white', bg='green', command=w.destroy)
outputLabel = Label(w, font=('Times New Roman', 14), fg='#B45253', bg='#FFE797', textvariable=displayResult)

# Place Widgets 
title.place(x=140, y=50)
gameInstruction.place(x=95, y=95) 
user_input.place(x=180, y=150) 
guess_button.place(x=370, y=147) 
exit_button.place(x=220, y=300) 
outputLabel.place(x=100, y=220)


w.mainloop()
```

```python
# What is Sentiment Analysis?
 
# Sentiment analysis is the process of analyzing digital text
# to determine if the emotional tone of the message is positive,
# negative, or neutral. Today, companies have large volumes of
# text data like emails, customer support chat transcripts,
# social media comments, and reviews. Sentiment analysis tools
# can scan this text to automatically determine the author‚Äôs
# attitude towards a topic. Companies use the insights from
# sentiment analysis to improve customer service and increase brand reputation.
 
# VADER
# VADER stands for Valence Aware Dictionary and Sentiment Reasoner.
# It‚Äôs a tool used for sentiment analysis, which is basically a
# way to figure out if a piece of text is expressing positive,
# negative, or neutral emotions.
 
# VADER doesn‚Äôt just say positive or negative,
# it assigns a score between -1 (most negative)
# and +1 (most positive) to show how strong the sentiment is.
 
# Research suggests VADER achieves the score of 0.96 for sentiment classification.
 
# The SentimentIntensityAnalyzer class provides a method called polarity_scores()
# that takes a piece of text as input and returns a dictionary containing
# the sentiment scores for the text.
# The dictionary contains four keys: neg, neu, pos, and compound.
```

```python
# Sentiment Analyzer App

from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sid_obj = SentimentIntensityAnalyzer()

w = Tk()
w.title('Sentiment Analyzer')
w.geometry('600x400')
# background color for window
w.config(bg='#F2AEBB')
w.resizable(width=False,height=False)

# main logic

displayResult = StringVar()

def check_sentiment():
    userInput = user_input.get()
    res = TextBlob(userInput)
    res = sid_obj.polarity_scores(res)

    if res['compound'] >= 0.5:
        displayResult.set('Positive Text')
        user_input.delete(0,END)
    elif res['compound'] <= -0.5:
        displayResult.set('Negative Text')
        user_input.delete(0,END)
    else:
        displayResult.set('Neutral Text')
        user_input.delete(0,END)
        
# Create Widgets
title = Label(w,text='Sentiment Analyzer Please Insert Some Text Below',
             font=('Arial',14),fg='#F5EDED',bg='#6482AD')
user_input = Entry(w,font=('Arial',14))
button1 = Button(w,text='Click Here To Analyze',font=('Arial',13),fg='#7C00FE',
                 bg='#BC9F8B',command=check_sentiment)
outputLabel = Label(w,font=('Arial',18),fg='#522258',bg='#6482AD',
                   textvariable=displayResult)

# Place Widgets
title.place(x=60,y=50)
user_input.place(x=150,y=120)
button1.place(x=390,y=120)
outputLabel.place(x=250,y=200)

w.mainloop()


```

```python
# Kpop Flashcards Quiz

from tkinter import *
import tkinter as tk
import random

flashcards = [
    {
        "question": "What does 'K-pop' stand for?",
        "correct": "Korean Pop",
        "options": ["Korean Rock", "Krazy Pop", "Kinetic Pop"]
    },
    {
        "question": "What is BLACKPINK‚Äôs official fandom name?",
        "correct": "BLINK",
        "options": ["BLACKIE", "BLUSH", "BLOOM"]
    },
    {
        "question": "What entertainment company is home to EXO?",
        "correct": "SM Entertainment",
        "options": ["YG Entertainment", "JYP Entertainment", "Big Hit"]
    },
    {
        "question": "What is the lightstick of BTS called?",
        "correct": "ARMY Bomb",
        "options": ["BTS Torch", "Light Bomb", "ARMY Stick"]
    },
    {
        "question": "What is Stray Kids‚Äô official fandom name?",
        "correct": "STAY",
        "options": ["STRAYS", "SKZEE", "SKIDZOO"]
    },
    {
        "question": "Who is the leader of Stray Kids?",
        "correct": "Bang Chan",
        "options": ["Hyunjin", "Lee know", "Changbin"]
    },
    {
        "question": "Which company manages LE SSERAFIM?",
        "correct": "Source Music",
        "options": ["Starship", "Big Hit", "SM Entertainment"]
    },
    {
        "question": "Which former IZ*ONE members are in LE SSERAFIM?",
        "correct": "Kim Chaewon and Sakura",
        "options": ["Wonyoung and Yujin", "Chaeyeon and Yena", "Minju and Nako"]
    }
]

class FlashCardApp:
    def __init__(self, kpop):
        self.kpop = kpop
        self.kpop.title("K-Pop Flash Card Quiz")
        self.kpop.geometry("450x500")
        self.score = 0
        self.total = 0
        self.limit = 5
        self.questions_used = []
        self.selected = False  # Defined up front

        # Start Screen 
        self.start_frame = tk.Frame(self.kpop, bg="#B3BFFF")
        self.start_frame.pack(fill="both", expand=True)

        self.title_label = tk.Label(self.start_frame, text="üé§ K-Pop Flash Card Quiz üé∂", font=("Arial", 18, "bold"), bg="#BADFDB")
        self.title_label.pack(pady=60)

        self.start_button = tk.Button(self.start_frame, text="Start", font=("Arial", 14), command=self.start_quiz)
        self.start_button.pack()

        #  Quiz Screen 
        self.quiz_frame = tk.Frame(self.kpop, bg="#FCB53B")
        
        self.question_label = tk.Label(self.quiz_frame, text="", font=("Arial", 14), wraplength=400, bg="#FCB53B")
        self.question_label.pack(pady=20)

        self.option_buttons = []
        for i in range(4):
            btn = tk.Button(self.quiz_frame, text="", font=("Arial", 12), width=30, command=lambda idx=i: self.check_answer(idx))
            btn.pack(pady=5)
            self.option_buttons.append(btn)

        self.feedback_label = tk.Label(self.quiz_frame, text="", font=("Arial", 12, "bold"), bg="#708993")
        self.feedback_label.pack(pady=10)

        self.next_button = tk.Button(self.quiz_frame, text="Next", font=("Arial", 12), command=self.next_card)
        self.score_label = tk.Label(self.quiz_frame, text="Score: 0 / 0", font=("Arial", 12), bg="#708993")

        # Final Result Screen 
        self.result_frame = tk.Frame(self.kpop, bg="#7D8D86")

        self.result_label = tk.Label(self.result_frame, text="", font=("Arial", 16, "bold"), bg="#7D8D86")
        self.result_label.pack(pady=40)

        self.restart_button = tk.Button(self.result_frame, text="Restart", font=("Arial", 12), command=self.restart_quiz)
        self.restart_button.pack(pady=10)

        self.exit_button = tk.Button(self.result_frame, text="Exit", font=("Arial", 12), command=self.kpop.destroy)
        self.exit_button.pack()

    def start_quiz(self):
        self.start_frame.pack_forget()
        self.quiz_frame.pack(fill="both", expand=True)
        self.score_label.pack(pady=5)
        self.next_card()

    def next_card(self):
        if self.total >= self.limit:
            self.show_result()
            return

        self.feedback_label.config(text="")
        self.selected = False
        self.next_button.pack_forget()

        # Avoid repeating same question
        while True:
            self.current_card = random.choice(flashcards)
            if self.current_card not in self.questions_used:
                self.questions_used.append(self.current_card)
                break

        all_options = self.current_card["options"] + [self.current_card["correct"]]
        random.shuffle(all_options)

        self.question_label.config(text=self.current_card["question"])
        self.correct_answer = self.current_card["correct"]
        self.current_options = all_options

        for i in range(4):
            self.option_buttons[i].config(
                text=all_options[i], state="normal", bg="SystemButtonFace"
            )

    def check_answer(self, index):
        if self.selected:
            return
        self.selected = True

        selected_answer = self.current_options[index]
        self.total += 1

        if selected_answer == self.correct_answer:
            self.score += 1
            self.feedback_label.config(text="‚úÖ Correct!", fg="green")
            self.option_buttons[index].config(bg="#90EE90")
        else:
            self.feedback_label.config(text=f"‚ùå Incorrect! Correct: {self.correct_answer}", fg="red")
            self.option_buttons[index].config(bg="#FBF3D1")
            for i, option in enumerate(self.current_options):
                if option == self.correct_answer:
                    self.option_buttons[i].config(bg="#90EE90")

        for btn in self.option_buttons:
            btn.config(state="disabled")

        self.score_label.config(text=f"Score: {self.score} / {self.total}")
        self.next_button.pack(pady=10)

    def show_result(self):
        self.quiz_frame.pack_forget()
        self.result_label.config(
            text=f"üéâ Quiz Complete!\n\nYour Score: {self.score} / {self.limit}"
        )
        
        self.result_frame.pack(fill="both", expand=True)

    def restart_quiz(self):
        self.result_frame.pack_forget()
        self.score = 0
        self.total = 0
        self.questions_used = []
        self.score_label.config(text="Score: 0 / 0")
        self.quiz_frame.pack(fill="both", expand=True)
        self.next_card()

# Run the app
kpop = tk.Tk()
app = FlashCardApp(kpop)
kpop.mainloop()

    
    
```
